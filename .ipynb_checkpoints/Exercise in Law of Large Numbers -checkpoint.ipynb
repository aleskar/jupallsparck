This is simply a way to demonstrate the LLN.  randn() returns a random number, with a normal distirubtion around zero, where the value 1 is one standard deviation away from 0.  So, we would expect that 68.2% of the time, the random number returned falls between -1 and 1.  [image source](https://upload.wikimedia.org/wikipedia/commons/8/8c/Standard_deviation_diagram.svg) <img src="images/Standard_deviation_diagram.svg">

We will test this assumption by running randn() a certain number of times, and then examine the percentage of times across those tests where the reutrned value is in fact between -1 and 1.  
We would assume that the result should be about 68.2% of the time (it returns value between -1 and 1).  However, the number of times randn() is run for the test can make a big difference.  We examine this difference by running the test experiement several times, with each time increasing the number of tests performed.

Overall steps:
- Testing the result of sampling randn(), **n** number of times
    - Test 1 n = 10
    - Test 2 n = 1000
    - Test 3 n = 10000


We are then going to run this testing of n samples for a certain number of **runs**.  I did this because it helps to point out the greater level of variation when n is small.  You'll see that with the 10 printed histograms where n=10, the distribution curves when compared graph to graph are 'all over the place'.  Whereas the 10 graphs produced when n=10000, all look very similar to each other. 

The following is a [Reference for fnx randn()](https://docs.scipy.org/doc/numpy-1.14.0/reference/generated/numpy.random.randn.html)

To access the randn() function we will need to
```python
import numpy as np
from numpy.random import randn
```

# Test 1: n = 10

````python
# Exercise in Law of Large Numbers
# Date: July 6, 2018
# Alex Skarulis
# Instructor: Kirill Eremenko

import numpy as np
from numpy.random import randn
import matplotlib.pyplot as plt
````

````python
# TEST 1 - n = 10
# runs we will set at 100 and leave.  This will be used to create graph showing distribution using exactly 10 '*'s 
runs = 10

# CHANGE n below to adjust number of samplings of randn
n = 10

# denominator of total times randn is sampled
totRuns = runs * n

# cumHits set to zero, but keeps running total of postive readings to a single overall average
cumHits = 0

print("This is a series of n tests running 'randn'.  In this test, n = ", n)
print("We expect the number of times randn returns a value between -1 and 1 to be 68.2%")
print("We run the", n, "test series", runs, "times to and create a simple text graph that can be used to compare the distribution of results with tests of difference size n.")

for run in range(runs):
    hits = 0
    a=[]
    for i in range(n):
        x = randn()
        if x > -1 and x < 1:
            hits = hits + 1
            cumHits = cumHits + 1
        a.append(x)
    pct = 100 * (hits / n)
    plt.hist(a,100)
    titl = "Histogram of randn() run " + str(n) + " times. $\mu=0$, $\sigma=1$ Run # " + str(run)
    plt.title(titl)
    plt.ylabel('Observations')
    plt.show()
    len(a)
    print("-------------------------------------------------------------------------")
    
print("========================================================================")
print("The overall average was", 100 * (cumHits/totRuns), "%. Expected average is: 68.2%")
````
